```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
sentences = nltk.sent_tokenize(para)
```

```python
stem = PorterStemmer()
lem = WordNetLemmatizer()
```

```python
import re

corpus = []

for i in sentences:
    clean = re.sub('[^a-zA-Z]', ' ', i)
    clean = clean.lower()
    corpus.append(clean)
```

```python
nltk.download('stopwords')
```

Stemming

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for j in words:
        if j not in set(stopwords.words('english')):
            print(stem.stem(j))
```

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for j in words:
        if j not in set(stopwords.words('english')):
            print(lem.lemmatize(j))
```

```python
para = """
Narendra Damodardas Modi (Gujarati: [ˈnəɾendɾə dɑmodəɾˈdɑs ˈmodiː] ⓘ; born 17 September 1950)[a] is an Indian politician serving as the current Prime Minister of India since 26 May 2014. Modi was the chief minister of Gujarat from 2001 to 2014 and is the Member of Parliament (MP) for Varanasi. He is a member of the Bharatiya Janata Party (BJP) and of the Rashtriya Swayamsevak Sangh (RSS), a right wing Hindu nationalist paramilitary volunteer organisation. He is the longest-serving prime minister outside the Indian National Congress.[4]

Modi was born and raised in Vadnagar in northeastern Gujarat, where he completed his secondary education. He was introduced to the RSS at the age of eight. At the age of 18, he was married to Jashodaben Modi, whom he abandoned soon after, only publicly acknowledging her four decades later when legally required to do so. Modi became a full-time worker for the RSS in Gujarat in 1971. The RSS assigned him to the BJP in 1985 and he rose through the party hierarchy, becoming general secretary in 1998.[b] In 2001, Modi was appointed Chief Minister of Gujarat and elected to the legislative assembly soon after. His administration is considered complicit in the 2002 Gujarat riots,[c] and has been criticised for its management of the crisis. According to official records, a little over 1,000 people were killed, three-quarters of whom were Muslim; independent sources estimated 2,000 deaths, mostly Muslim.[13] A Special Investigation Team appointed by the Supreme Court of India in 2012 found no evidence to initiate prosecution proceedings against him.[d] While his policies as chief minister were credited for encouraging economic growth, his administration was criticised for failing to significantly improve health, poverty and education indices in the state.[e]
"""
```

```python
sent = nltk.sent_tokenize(para)
```

Corpus -> Sentence -> Words <br />
Lemmatize + remove stopwords

```python
corpus = []

for i in sent:
    cleaned = re.sub('[^a-zA-Z]', ' ', i)
    cleaned = cleaned.lower()
    cleaned = cleaned.split()
    lemmatized = [lem.lemmatize(word) for word in cleaned if word not in set(stopwords.words('english'))]
    lemmatized = ' '.join(lemmatized)
    corpus.append(lemmatized)
```

```python
corpus
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = cv.fit_transform(corpus)
```

```python
cv.vocabulary_
```

```python
X[0].toarray()
```

NGrams

```python
cv = CountVectorizer(ngram_range=(2,3)) # bigrams and trigrams
Y = cv.fit_transform(corpus)
```

```python
cv.vocabulary_
```

```python
Y[0].toarray()
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer() # can also pass ngram_range here
Z = tfidf.fit_transform(corpus)
```
